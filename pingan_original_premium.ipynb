# 数据整合

```python
def get_year_original_premium(year):
    #读取excel文件
    df = pd.read_excel(r'data/pingan/平安原保费.xlsx', sheet_name=year)
    #先删除空行后删除空列
#     df = df.dropna(axis=1, how='all')
    df = df.dropna(axis=0, how='all').dropna(axis=1, how='all')
    #转置时间为索引
    df = df.T
    #设置月份时间索引
    dates = pd.date_range(year+'-01-31',year+'-12-31',freq='M')[:len(df)]
    index = dates.to_period('M')
    df = df.set_index(index)
    return df
```

```python
df_2016 = get_year_original_premium('2016')
df_2016
```

```python
df_2017 = get_year_original_premium('2017')
df_2017
```

```python
df_2018 = get_year_original_premium('2018')
df_2018
```

```python
dfs = [get_year_original_premium(i) for i in ['2016', '2017', '2018']]
```

```python
df = pd.concat(dfs, axis=1)
df
```

```python
def get_company(df):
    return df.iloc[:,:4]
```

```python
def get_life_insurance(df):
    return df.iloc[:,4:-4]
```

```python
def get_asset_insurance(df):
    return df.iloc[:,-4:]
```
